```python
from ampligraph.latent_features import ComplEx
import pandas as pd
import numpy as np
```

```python
triplets = pd.read_csv('../data/wikipages_triplets.csv')
```

```python
triplets.head()
```

```python
triplets.shape
```

```python
wikipages_cleaned = pd.read_csv('../data/wikipages_cleaned.csv')
```

```python
wikipages_cleaned.head()
```

```python
wikipages_cleaned.views.describe()
```

```python
popular_wikipages = wikipages_cleaned.sort_values(by = ['views'], ascending = False).iloc[:450000]['wikidata_numeric_id'].unique()
```

```python
popular_triplets = triplets[(triplets.source_item_id.isin(popular_wikipages)) \
                            & (triplets.target_item_id.isin(popular_wikipages))]
```

```python
popular_triplets.shape
```

```python
X = popular_triplets[['source_item_id', 'edge_property_id', 'target_item_id']].values
```

```python
model = ComplEx(optimizer = 'sgd', k = 100, loss = 'pairwise', epochs = 1, )
```

```python
model.fit(X)
```
